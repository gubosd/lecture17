## 머신러닝
# 상관계수
- Pearson correlation coefficient

$$ correlation = {1 \over N} \sum_i {(x_i - \bar x) (y_i - \bar y) \over {\sigma_x \sigma_y}} $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
```

```python
len(cancer.feature_names)
```

```python
for i in range(30):
    print('%02d : %s' % (i, cancer.feature_names[i]))
```

```python
corr = np.corrcoef(cancer.data.T) # 속성이 행으로 되어야 함
corr.shape
```

```python
plt.figure(figsize=[10,10])
plt.imshow(corr)
plt.colorbar(shrink=0.7)
plt.xticks(range(30),cancer.feature_names,rotation=90,ha='center')
pass
```
